```python
d[['Cost', 'Profit']].cov()   # Covariance matrix
```

The result here is a covariance matrix for the Cost and Profit columns of the dataset. Covariance measures how much two variables change together.

**What is Covariance?**
**Covariance** is a measure of the relationship between two variables and how they vary together.
- **Positive covariance** means that when one variable increases, the other tends to increase as well.
- **Negative covariance** means that when one variable increases, the other tends to decrease.
  
However, unlike correlation, covariance does not have a fixed range (like -1 to 1), so the scale of the numbers depends on the units of the data.

**Explanation of the Matrix:**
1. **Cost vs. Cost (786,718.68):**
    - This is the covariance of Cost with itself, which is simply the variance of Cost. It tells you how much the cost values spread around their mean.
2. **Profit vs. Profit (206,890.92):**
    - This is the covariance of Profit with itself, which is the variance of Profit. It tells you how much the profit values spread around their mean.
3. **Cost vs. Profit (363,948.05):**
   - This is the **covariance between Cost and Profit.** Since this value is positive and relatively large, it suggests that **Cost and Profit tend to increase or decrease together** â€”as one rises, so does the other.
    - The large magnitude here indicates that thereâ€™s a strong relationship, but since covariance is dependent on the units of the variables, itâ€™s not as standardized as correlation. However, a positive value is consistent with the strong positive correlation seen in the correlation matrix.
      
**Business Implications:**
- **Cost and Profit Relationship:** The positive covariance confirms that thereâ€™s a relationship where higher costs are associated with higher profits. It reinforces the idea that larger or more expensive sales (higher costs) are bringing in more profit.

- **Magnitude of Variability:** The higher variance of Cost (786,718.68) compared to Profit (206,890.92) suggests that the costs fluctuate more widely than profits, meaning cost variability is higher than profit variability.

While covariance tells us thereâ€™s a relationship, the exact strength and direction are better understood using correlation.

**Group Analysis: groupby() and Aggregations**

```python
pp = d.groupby('Product')['Profit'].sum().sort_values(ascending = False)
```

```python
print(pp.reset_index().to_string())
```

```python
pp.head(10)
```

1. d.groupby('Product'):
This groups the DataFrame d by the **'Product'** column. Each unique product in the dataset will have its own group containing all the rows related to that product.
2. ['Profit'].sum():
For each product group, the code calculates the **sum** of the profits. This gives the **total profit** generated by each product across all transactions.
3. sort_values(ascending=False):
The total profits are then sorted in **descending order** (from highest to lowest). The product that generated the highest profit will be listed first, and the one with the lowest profit will be last.
4. pp.head(10):
Finally, the .head(10) method returns the **top 10 products** that generated the most profit.

**What This Means:**
- The code is determining the top 10 most profitable products by summing up the profit for each product and then sorting them in descending order.
- The resulting output will display the product names and their corresponding total profits.

**Business Implication:**
- **Identify Best Sellers:** This is useful for identifying which products are driving the most profit for the business. It helps focus on the products that contribute the most to overall profitability.
- **Decision Making:** With this information, you can make decisions about stock, marketing, or even expanding the product line based on which products are performing the best financially.

```python
# Outliers: Calculate Q1 and Q3

Q1 = d['Profit'].quantile(0.25)
Q3 = d['Profit'].quantile(0.75)
IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

```python
# Filter outliers

outliers = d[(d['Profit'] < (Q1 - 1.5 * IQR)) | (d['Profit'] > (Q3 + 1.5 * IQR))]
```
